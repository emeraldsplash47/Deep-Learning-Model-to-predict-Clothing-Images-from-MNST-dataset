```python
# import tensoflow and numpy
import tensorflow as tf
import numpy as np

# load the Fashion-MNIST dataset to train as well as test our data
image_data = tf.keras.datasets.fashion_mnist
```

```python
# assign labels to our test and train data
(train_images, train_labels), (test_images, test_labels) = image_data.load_data()


# plotting our data to verfiy if our data has been loaded successfully
import matplotlib.pyplot as plt
plt.imshow(test_images[8])
```

```python

# normalizing our data (pre-processing)
# remapping our values to between 0 to 1
train_images = train_images/ 255.0
test_images = test_images/ 255.0


np.max(train_images)
```

```python
# defining our model, and layers 
# our input shape is 28* 28
# our activation function is "relu"
# we use a dense layer (fully connected layer)
# our output layer has 10 nodes

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128,activation="relu"),
    tf.keras.layers.Dense(10)
])


# we use the "adam" optimizer
model.compile(optimizer= "adam", 
              loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics = ["accuracy"]
              )

# number of epochs = 10
# 10 iterations of training this model
model.fit(train_images,train_labels,epochs= 10)

```

```python
# testing our model
model.predict(test_images)[0]
test_labels[0]
```
